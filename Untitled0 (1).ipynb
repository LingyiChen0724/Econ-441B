```python
import pandas as pd
from google.colab import drive
import matplotlib.pyplot as plt

drive.mount('/content/gdrive/', force_remount= True)
```

```python
df = pd.read_csv("/content/gdrive/MyDrive/Econ 441B/^VIX.csv")
```

```python
df.set_index("Date", inplace = True)
df.index = pd.to_datetime(df.index)
```

```python
plt.figure(figsize = (12,5))
plt.plot(df["Close"])
plt.title("VIXCLS INDEX")
plt.xlabel("Date")
plt.ylabel("Vix")
plt.grid()
plt.show()
```

VIX index has hugh fluctuations from 2022 to 2023. It is around 21 in Jan 1st 2022. It starts to increase and peaks at about 32, following a hugh drop in Feb 2022. The index reaches its highest value in March 2022 around 36. And then it starts to decrease again. Recently, the index is decreasing and reaches a low value around 19.
